```python
def group_activity (activity_name):
    keywords1 = ['road', 'pcc', 'construction', 'cc','bridge','ROAD REPAIRING']
    keywords2 = ['drinking water', 'mark ll', 'tubewell', 'sinking', 'ro']
    keywords3 = ['solar','electric','rooftop','green','renewable','light']
    keywords4 = ['building']
    keywords5 = ['children park','beautification']
    keywords6 = ['drain','nala']
    keywords7 = ['community toilet','latrin']
    
    for i in keywords1 :
        if i in activity_name.lower():
            return 'Road'
    for i in keywords2 :
        if i in activity_name.lower():
            return 'Drinking Water'
        
    for i in keywords3 :
        if i in activity_name.lower():
            return 'Solar Elctrification'     
        
    for i in keywords4 :
        if i in activity_name.lower():
            return 'Building'   
        
    for i in keywords5 :
        if i in activity_name.lower():
            return 'Children Park'   
        
    for i in keywords6 :
        if i in activity_name.lower():
            return 'Drain'    
        
    for i in keywords7 :
        if i in activity_name.lower():
            return 'Community Toilet' 
    else:
        return 'other'
        
new_df['Activity Group'] = new_df['Activity Name'].apply(group_activity)
print(new_df)
grouped_df = new_df.groupby(['GP Name','Activity Group'])['Estimated Cost'].agg(['sum', 'count']).reset_index()
grouped_df.rename(columns={'count': 'no. of schemes'}, inplace=True)
grouped_df
  
    
# grouped_df = df.groupby(['GP Name', 'Activity Group'])['Estimated Cost'].agg(['sum', 'count']).reset_index()
# grouped_df.rename(columns={'count': 'no. of schemes'}, inplace=True)
# grouped_df
        
```

```python

#monkey patching
def div (x,y):
    return x/y
def custom_div (x,y):
    if y == 0:
        return  float('inf')
    
div = custom_div
print(div(4,0))
```

```python
#decorator

def doubled_no(func):
    def wrapper(*arg,**kwarg):
        return func(*arg,**kwarg)*2
    return wrapper
@doubled_no
def add(x,y):
    return x+y
@doubled_no
def substract(x,y):
    return x-y
print(add(3,4))
print(substract(8,2))
```

```python
def largest(n):
    lar = float('-inf')
    for i in n:
        if i > lar:
            lar = i
    return lar
largest([3,4,5,6,7,87])
```

```python
def second_largest(n):
    lar = float('-inf')
    sec_lar = float('-inf')
    for i in n:
        if i > lar:
            sec_lar = lar
            lar = i
        elif i > sec_lar and i != lar:
            sec_lar = i

    return sec_lar
second_largest([3,5,4,1,7,8,6,9,45,44])
```

# Manipulation on string by .translate() / .maketrans() 
str.maketrans('@#$', ' ') creates a translation table that maps @, #, and $ to spaces.

str.translate(my_table) applies this translation table to replace these special characters with spaces in the string.

```python
str  = 'Poulomi'
mytable = str.maketrans('Po','tf','mi')  # x replaced by y, z will remove
print(str.translate(mytable))
```

'@#$'  Removed

```python
str = 'soumik das 12345 @#$'
my_table = str.maketrans(' ' , ' ','@#$')
str.translate(my_table)
```

'@#$' replaced by '   '

```python
str = 'soumik das 12345 @#$'
my_table = str.maketrans('@#$', '   ')
print(str.translate(my_table))
```
